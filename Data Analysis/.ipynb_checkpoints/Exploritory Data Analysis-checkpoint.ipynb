```python
data.Installs = data.Installs.str.replace('+', '')
data.Installs = data.Installs.str.replace(',','')
data.Installs = data.Installs.apply(lambda x: int(x))

data['Size'].replace('Varies with device', np.nan, inplace = True ) 
data.Size = (data.Size.replace(r'[kM]+$', '', regex=True).astype(float))
data['Size'].fillna(data.groupby('Category')['Size'].transform('median'),inplace = True)
```

### I decided to clean the installs values like in the famous_apps dataset and decided to replace the 'Varies with device' values with NaNs and applying median of size for of apps in each category transformation 

```python
data.Price.value_counts()
```

```python
data['Price'] = data.Price.str.replace('$', '')
data['Price'] = data.Price.apply(lambda x: float(x))
```
