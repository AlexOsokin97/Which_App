```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_apps = pd.read_csv('EDA_df.csv')
```

```python
df_apps.head()
```

```python
df_apps.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_apps.head()
```

```python
df_apps.info()
```

```python
df_apps.describe()
```

```python
df_apps.dropna(inplace=True)
```

```python
df_apps.info()
```

```python
df_apps.describe()
```

```python
df_apps[['Category','Type','Content Rating']] = df_apps[['Category','Type','Content Rating']].astype('category')
df_copy = df_apps.copy()
cat_columns = df_copy[['Type','Content Rating']].columns
df_copy[cat_columns] = df_copy[cat_columns].apply(lambda x: x.cat.codes)
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(10, 8)
corr = df_copy.corr()
cmap = 'coolwarm'
sns.heatmap(corr, cmap=cmap, annot=True, ax=ax)
fig.savefig('corrHeatmap.png')
```

```python
df_copy['Category'].value_counts()
```

```python
fig1, ax1 = plt.subplots(figsize=(15,8))
x = df_copy['Category']
y1 = df_copy['Installs']
sns.barplot(x=x, y=y1, palette="rocket",ci=None, ax=ax1)
ax1.set_ylabel('Avg. Installs (Millions)')
ax1.set_title('Avg. Installs per Category')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=40, ha="right")
fig1.savefig('AvgInstallperCategory.png')
```

```python
fig2, ax2 = plt.subplots(figsize=(15,8))
x = df_copy['Category']
y1 = df_copy['Size']
sns.barplot(x=x, y=y1, palette="rocket",ci=None, ax=ax2)
ax2.set_ylabel('Avg. App Size (Mb)')
ax2.set_title('Avg. App Size per Category')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=40, ha="right")
```

```python
fig3, ax3 = plt.subplots(figsize=(10,8))
x = df_copy['Reviews']
y1 = df_copy['Installs']
sns.regplot(x=x, y=y1)
```
